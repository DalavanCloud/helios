```python
def class_filler(array,value,iterations):
    "array: array to store values"
    "value: value to be store"
    "iterations: # of iterations to be stored"
    for _ in range(0,iterations):
        array.append(value)
```

```python
import os

mfcc_class = []

for i in range(0,8):
    path = 'train_waves_svm\\female' + str(i+1)
      
    for filename in os.listdir(path):
        x, fs = librosa.load(path + '\\' + filename)
        
        if i == 0: #Inititalize mfcc_data array
            mfcc_data = librosa.feature.mfcc(x, sr=fs)
            mfcc_data = mfcc_data.transpose()
            class_filler(mfcc_class,(i+1),mfcc_data.shape[0])
            break
        
        features = librosa.feature.mfcc(x, sr=fs)
        features = features.transpose()
        
        class_filler(mfcc_class,(i+1),features.shape[0])
        
        mfcc_data = np.vstack((mfcc_data,features)) #Stack mfcc matrices
        
        print filename
        print features.shape
        
        
```

```python
#NUMPYfy things

mfcc_data= np.asarray(mfcc_data) #Convert mfcc_data to an NUMPY
print type(mfcc_data)
print mfcc_data.shape

mfcc_class = np.asarray(mfcc_class)
print mfcc_class.shape
```

## SVM

```python
from sklearn import svm
svm_model = svm.SVC(kernel='rbf', gamma=0.7, C=1)
#svm_model = svm.SVC(kernel='poly', degree=2, C=1)
#Training
svm_model.fit(mfcc_data[:,:],mfcc_class)
```

```python
#Testing

prediction_list = []

for i in range(0,8):
    #path = 'test_waves_svm\\female' + str(i+1)
    path = 'train_waves_svm\\female' + str(i+1)
    
    for filename in os.listdir(path):
        predictions = []
        x, fs = librosa.load(path + '\\' + filename)
        mfcc_result = librosa.feature.mfcc(x, sr=fs)
        mfcc_result = mfcc_result.transpose()
        mfcc_result = np.asarray(mfcc_result)
        
        #Predict and save
        for row in mfcc_result:
            #print row
            #print row.shape
            prediction = svm_model.predict(row.reshape(1,-1))
            print prediction
            predictions.append(prediction)
            
        prediction_list.append(predictions)
    
```
